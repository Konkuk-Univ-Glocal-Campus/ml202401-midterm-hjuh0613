```python
%pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#지수표현식 없애기 
pd.options.display.float_format = '{:.5f}'.format

data = pd.read_csv("./서울시 상권분석서비스(집객시설-상권).csv", encoding='euc-kr')
data.head()
```

```python
data.corr()
```

```python
data = data.fillna(0)
data = data[(data['기준_년분기_코드'] == 20231)]
data2 = data[['상권_코드_명','극장_수','은행_수']]
data2.head()
```

```python
X = data2.iloc[:,[1,2]].values #x값으로 극장수, 은행 수 지정
m = X.shape[0] #전체 데이터 개수
n = X.shape[1] #특징 개수(극장수, 은행수)
```

```python
X = data2.iloc[:,[1,2]].values #x값으로 극장수, 백화점 수 지정
plt.scatter(X[:,0], X[:,1], c='black', label='상권')
plt.xlabel('극장 수')
plt.ylabel('은행 수')
plt.legend()
plt.title('주요 집객 시설')
plt.show
```

```python
from sklearn.cluster import KMeans
inertia = []

for k in range(1,11):
    kmeans = KMeans(n_clusters=k).fit(X)
    value = kmeans.inertia_
    inertia.append(value)

pd.DataFrame(inertia).plot()
```

```python
kmeans = KMeans(n_clusters=4).fit(X) #군집 4개로 나누겠다
data2['cluster_id'] = kmeans.labels_
data2
```

```python
sns.lmplot(x='극장_수', y='은행_수', data=data2, fit_reg=False, scatter_kws={"s":150}, hue="cluster_id")
plt.title("은행과 극장수로 찾는 번화가")
```
